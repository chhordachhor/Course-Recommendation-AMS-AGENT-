$\text{Ploting Course Rating to get an overview of the distribution}$

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

df["Course Rating"] = pd.to_numeric(df["Course Rating"], errors='coerce')

median_rating = df["Course Rating"].median()
df["Course Rating"] = df["Course Rating"].fillna(median_rating)

ratings = df["Course Rating"]
ratings = ratings[ratings > 0]

if ratings.empty:
    raise ValueError("No valid ratings available after filtering!")

shape, loc, scale = stats.gamma.fit(ratings, floc=0)

print(f"Shape: {shape:.4f}, Loc: {loc:.4f}, Scale: {scale:.4f}")

sns.set(rc={'figure.figsize': (10, 6)})
fig, ax = plt.subplots()
sns.histplot(ratings, color='cyan', kde=True, stat="density", ax=ax)

x = np.linspace(ratings.min(), ratings.max(), 1000)
p = stats.gamma.pdf(x, shape, loc, scale)

ax.plot(x, p, 'r', label="Gamma Fit", linewidth=2)
ax.set_title("Course Rating Distribution", fontsize=18)
ax.set_xlabel("Course Rating", fontsize=14)
ax.set_ylabel("Density", fontsize=14)
ax.legend()

plt.show()
```

$\textbf{Generating Review Sentiments using Vader Sentiment}$

```python
def sentiment_analyzer(sentence):
    score = analyser.polarity_scores(sentence)
    return score
```

```python
neg=[]
neu=[]
pos=[]
comp=[]
for review in data1.reviews:
    scores = sentiment_analyzer(str(review))
    neg.append(scores['neg'])
    pos.append(scores['pos'])
    neu.append(scores['neu'])
    comp.append(scores['compound'])
```

$$\textbf{Review Analysis}$$

```python
data1['s_pos'] = pos
data1['s_neu'] = neu
data1['s_neg'] = neg
data1['s_comp'] = comp
```

```python
data1.head()
```

```python
for i in range(6,11):
    print ("----------------------------------------")
    print("Review : ",data1['reviews'][i])
    print("Positivity:",data1['s_pos'][i])
    print("Negativity:",data1['s_neg'][i])
    print ("----------------------------------------")
```
